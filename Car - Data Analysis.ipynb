```python
#importing libraries and reading the file
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)
df.head(n=5)
```

```python
#assign appropriate headers to the new variable headers
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base",
"length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio",
"horsepower","peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head(5)
```

```python
#To find all the datatypes
df.dtypes
```

```python
#replace '?' to NaN values
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
#Handling Missing Values
#To count all the missing values in the file
missing_value = df.isnull()
missing_value.head(5)
```

```python
#Calculate the mean of the value of the columns to use it to replace the null values
avg_norm_losses = df['normalized-losses'].astype("float").mean(axis=0)
avg_bore = df['bore'].astype("float").mean(axis=0)
avg_peak_rpm = df['peak-rpm'].astype("float").mean(axis=0)
avg_stroke = df['stroke'].astype('float').mean(axis = 0)
avg_horsepower = df['horsepower'].astype('float').mean(axis = 0)
print ("Average of normalised-losses =", avg_norm_losses)
print ("Average of bore =", avg_bore)
print ("Average of peak-rpm =", avg_peak_rpm)
print("Average of stroke:", avg_stroke)
print("Average of Horsepower:", avg_horsepower)
```

```python
#Replace NaN by the mean value calculated
df['normalized-losses'].replace(np.nan,avg_norm_losses, inplace= True)
df['bore'].replace(np.nan,avg_bore, inplace= True)
df['peak-rpm'].replace(np.nan,avg_peak_rpm,inplace= True)
df['stroke'].replace(np.nan,avg_stroke,inplace = True)
df['horsepower'].replace(np.nan,avg_horsepower, inplace=True)
```

```python
#drop the NaN values from price column
df.dropna(subset=["price"], axis= 0, inplace = True)
```

```python
#to replace columns with the most repeated values
num_of_door_freq = df['num-of-doors'].mode()
print ("Most frequently occuring number of doors is:",num_of_door_freq)
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
#Convert columns to the standard unit for easy comparision
df["city-mpg"] = 235/df ["city-mpg"]
df.rename(columns = {"city-mpg" : "city-L/100Km"} , inplace = True)
```

```python
df["highway-mpg"] = 235/df ["highway-mpg"]
df.rename(columns = {"highway-mpg" : "highway-L/100Km"} , inplace = True)
df.head(5)
```

```python
#convert to appropriate datatypes
df["price"] = df["price"].astype(int)
df["horsepower"] = df["horsepower"].astype(float)
df["peak-rpm"] = df["peak-rpm"].astype(float)
df["stroke"] = df["stroke"].astype(float)
df["bore"] = df["bore"].astype(float)
df["normalized-losses"] = df["normalized-losses"].astype(int)
```

```python
df.dtypes
```

```python
#data normalization. To scale the columns to range from 0 to 1
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
#Binning
bins= np.linspace(min(df["price"]),max(df["price"]),4)
group_names= ['Low','Medium','High']
df["price-binned"] = pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
bins
```
